```python
import numpy as np

def generate_gaussian_wigner_matrix(n):
    # Generate random matrix elements from a Gaussian distribution
    real_part = np.random.normal(0, 1, size=(n, n))
    imag_part = np.random.normal(0, 1, size=(n, n))
    
    # Create the upper triangular Wigner matrix
    wigner_upper = real_part + 1j * imag_part
    
    # Generate the full symmetric Wigner matrix
    wigner_full = np.triu(wigner_upper, k=1) + np.triu(wigner_upper, k=1).conj().T
    
    return wigner_full

# Generate a 5x5 Gaussian Wigner matrix
n = 5
gaussian_wigner_matrix = generate_gaussian_wigner_matrix(n)
print("Gaussian Wigner Matrix:")
print(gaussian_wigner_matrix)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_gaussian_wigner_matrix(n):
    # Generate random matrix elements from a Gaussian distribution
    real_part = np.random.normal(0, 1, size=(n, n))
    imag_part = np.random.normal(0, 1, size=(n, n))
    
    # Create the upper triangular Wigner matrix
    wigner_upper = real_part + 1j * imag_part
    
    # Generate the full symmetric Wigner matrix
    wigner_full = np.triu(wigner_upper, k=1) + np.triu(wigner_upper, k=1).conj().T
    
    return wigner_full

# Generate a Gaussian Wigner matrix
n = 100
gaussian_wigner_matrix = generate_gaussian_wigner_matrix(n)

# Compute all eigenvalues of the Gaussian Wigner matrix
eigenvalues = np.linalg.eigvalsh(gaussian_wigner_matrix)

# Plot the histogram of eigenvalues
plt.hist(eigenvalues, bins=50, density=True, edgecolor='black')
plt.title("Histogram of Eigenvalues")
plt.xlabel("Eigenvalues")
plt.ylabel("Density")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_diagonal_entries(n, distribution):
    if distribution == 'normal':
        diagonal_entries = np.random.normal(0, 1, n)
    elif distribution == 'uniform':
        diagonal_entries = np.random.uniform(-np.sqrt(3), np.sqrt(3), n)  # Uniform distribution with variance 1
    elif distribution == 'laplace':
        diagonal_entries = np.random.laplace(0, 1 / np.sqrt(2), n)  # Laplace distribution with scale parameter 1/sqrt(2)
    else:
        raise ValueError("Invalid distribution. Choose from 'normal', 'uniform', or 'laplace'.")
    
    return diagonal_entries

# Define the size of the matrix and the distribution for diagonal entries
n = 100
distribution = 'uniform'  # Change the distribution here ('normal', 'uniform', or 'laplace')

# Generate diagonal entries using the specified distribution
diagonal_entries = generate_diagonal_entries(n, distribution)

# Plot the histogram of diagonal entries
plt.hist(diagonal_entries, bins=20, edgecolor='black')
plt.title(f"Histogram of Diagonal Entries ({distribution} distribution)")
plt.xlabel("Diagonal Entries")
plt.ylabel("Frequency")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def generate_off_diagonal_entries(n, distribution):
    if distribution == 'normal':
        off_diagonal_entries = np.random.normal(0, 1, (n, n))
    elif distribution == 'uniform':
        off_diagonal_entries = np.random.uniform(-np.sqrt(3), np.sqrt(3), (n, n))  # Uniform distribution with variance 1
    elif distribution == 'laplace':
        off_diagonal_entries = np.random.laplace(0, 1 / np.sqrt(2), (n, n))  # Laplace distribution with scale parameter 1/sqrt(2)
    else:
        raise ValueError("Invalid distribution. Choose from 'normal', 'uniform', or 'laplace'.")
    
    # Set diagonal entries to zero
    np.fill_diagonal(off_diagonal_entries, 0)
    
    return off_diagonal_entries

# Define the size of the matrix and the distribution for off-diagonal entries
n = 100
distribution = 'uniform'  # Change the distribution here ('normal', 'uniform', or 'laplace')

# Generate off-diagonal entries using the specified distribution
off_diagonal_entries = generate_off_diagonal_entries(n, distribution)

# Flatten off-diagonal entries for histogram plot
off_diagonal_entries_flat = off_diagonal_entries.flatten()

# Plot the histogram of off-diagonal entries
plt.hist(off_diagonal_entries_flat, bins=20, edgecolor='black')
plt.title(f"Histogram of Off-Diagonal Entries ({distribution} distribution)")
plt.xlabel("Off-Diagonal Entries")
plt.ylabel("Frequency")
plt.show()
```
